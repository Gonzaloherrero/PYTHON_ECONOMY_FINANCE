```python
# b
# Parametros:
    
Capacidad_equipaje =3000 
Capacidad_asientos = 90
Pesomax_F = 50
Pesomax_nF = 20
Precio_F = 10000
Precio_nF = 6000
Aminimo_nF = 10
```

```python
import scipy.optimize as sco
import numpy as np
```

```python
z = np.array([Precio_F,Precio_nF])
a_ub= np.array([[1 , 1] , [Pesomax_F , Pesomax_nF] , [-1 , 0],[0,-1],[0,-1]])
b_ub=np.array([Capacidad_asientos , Capacidad_equipaje, 0 , 0, Aminimo_nF])
```

```python
resultado = sco.linprog(-z , A_ub=a_ub , b_ub=b_ub)
resultado
```

```python
A_F=round(resultado.x[0])
A_nF=round(resultado.x[1])
Beneficio=-round(resultado.fun)
```

```python
print(f"Para maximizar el beneficio la oferta de asientos para fumadores debe ser de: {A_F} asientos y de {A_nF} asientos para no fumadores, obteniendo asi un beneficio de {Beneficio} pesos")
```

## Programacion no Lineal

```python
import scipy.optimize as sco
import numpy as np
```

```python
def objetivo(x):
    return -(2*x[0]**2 + x[1]**2 + 2*x[2]**2 - 10*x[0] - 2*x[1] -3*x[2])
```

```python
def restriccion1(x):
    return -x[0] + 0.4 * sum(x)

def restriccion2(x):
    return -x[1] + 0.4 * sum(x)

def restriccion3(x):
    return -x[2] + 0.4 * sum(x)

def restriccion4(x):
    return -2*x[0] - 5*x[1] - 3*x[2] + 14000
```

```python
limites = [(0, 14000/2), (0, 14000/5), (0, 14000/3)]
```

```python
restricciones = [{'type': 'ineq', 'fun': restriccion1},{'type': 'ineq', 'fun': restriccion2},{'type': 'ineq', 'fun': restriccion3},{'type': 'eq', 'fun': restriccion4}]
```

```python
inicio=[0.0001,0.0001,0.0001]
```

```python
resultado =sco.minimize(objetivo, inicio, method='SLSQP', bounds=limites, constraints=restricciones)
```

```python
print(resultado)
```

```python
print('Litros de jugo A:', resultado.x[0])
print('Litros de jugo B:', resultado.x[1])
print('Litros de jugo C:', resultado.x[2])
print('Beneficio máximo:', -resultado.fun)
```

## Simulacion de Montecarlo

```python
import numpy as np
import  pandas as pd
```

```python
np.random.seed(100)
```

```python
x0 = []
x1 = []
x2 = []
```

```python
litros_jugo_puro=np.array([2,5,3])
```

```python
x0_max=14000/litros_jugo_puro[0]
x1_max=14000/litros_jugo_puro[1]
x2_max=14000/litros_jugo_puro[2]
```

```python
for i in range(100000):
    xx0=np.random.randint(0,x0_max)
    xx1=np.random.randint(0,x1_max)
    xx2=np.random.randint(0,x2_max)
    x0.append(xx0)
    x1.append(xx1)
    x2.append(xx2)
    
```

```python
result=list(zip(x0,x1,x2))
```

```python
result=pd.DataFrame(result)
```

```python
result.columns=["LA","LB","LC"]
```

```python
result
```

```python
# xi <= 0,4*sum(x)
result["produccion_total"]=result.LA+result.LB+result.LC
result
```

```python
result["% LA"]=(result.LA/result.produccion_total)*100
result["% LB"]=(result.LB/result.produccion_total)*100
result["% LC"]=(result.LC/result.produccion_total)*100
result
```

```python
filtro=result[(result["% LA"]<=40)&(result["% LB"]<=40)&(result["% LC"]<=40)]
filtro
```

```python
costo_x0=10
costo_x1=2
costo_x2=3
```

```python
filtro["ingresos"]=2*filtro.LA**2 + filtro.LB**2 + 2*filtro.LC**2
filtro["costos"]=costo_x0*filtro.LA + costo_x1*filtro.LB + costo_x2*filtro.LC
filtro
    
```

```python
filtro["beneficios"]=filtro.ingresos-filtro.costos
filtro
```

```python
filtro.loc[filtro.beneficios.idxmax()]
```

```python
Max_Beneficios=round(filtro.loc[filtro.beneficios.idxmax()])
Max_Beneficios
```
